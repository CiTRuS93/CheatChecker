# CheatChecker
## using bert pretrained module and cosine similarity to find possible cheating in exam.

```python
from bertAnalyzer import bertAnalyzer
analyzer =  bertAnalyzer()

analyzer.set_data_path('D:\music\בדיקה-שאלות-תגובות.csv') #path to moodle's exported answers
analyzer.load_data()
df =  analyzer.analyze_data()
```

```python
df['mean'] = df.mean(axis = 1)
df.sort_values(by='mean',ascending=False).head(50)
```

### in this experiment we had 20 examinees and 2 pair of cheaters
#### as we can see the cheater couples are well separated from all the other examinees

```python
%matplotlib notebook
from matplotlib.pyplot import *
fig, ax = subplots()
df[df['mean']<0.4]['mean'].plot.hist(bins=12, alpha=0.5)
df[df['mean']>0.50]['mean'].plot.hist(bins=3, alpha=1)
ax.legend(["Non-suspects", "Suspects"]);
ax.set_xlim(-0.02,0.8);
ax.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off

# ax.annotate('', xy=(0.67, 1), xytext=(0.45, 10),
#             arrowprops=dict(facecolor='black', shrink=0.05),
#             )
```

```python
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """

    color = 'red' if isinstance(val, float) and val > 0.5 else ''
    
    return 'background-color: %s' % color
```
